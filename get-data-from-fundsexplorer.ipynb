```python
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.chrome.service import Service
from selenium import webdriver
import pandas as pd
import time
```

```python
path = 'chromedriver-win64.exe'

options = Options()
options.headless = True
options.add_argument("--headless=new")

service = Service(executable_path=path)
driver = webdriver.Chrome(service=service, options=options)
```

```python
driver.get('https://www.fundsexplorer.com.br/funds')
time.sleep(5)
```

```python
containers = driver.find_elements("xpath", "//div[@class='tickerBox link-tickers-container ']")
```

```python
fii_links = []
for container in containers:
    attribute = container.get_attribute('onclick')
    attribute = attribute.replace("location.href='", '')
    attribute = attribute.replace("';", '')
    fii_links.append(attribute)
```

```python
def get_value_by_xpath(path):
    try:
        item = driver.find_element("xpath", path)
        return item.text
    except:
        return '0'
```

```python
data = []
for index,link in enumerate(fii_links):
    driver.get(link)
    time.sleep(7)

    fii_data = {
        'Nome': get_value_by_xpath('//*[@id="tickerName"]'),
        'Preco': get_value_by_xpath('/html/body/div[4]/section[1]/div/div/div[1]/div[1]/p'),
        'Liquidez Média Diária': get_value_by_xpath('/html/body/div[4]/section[2]/div/div[1]/p[2]/b'),
        'Último Rendimento': get_value_by_xpath('/html/body/div[4]/section[2]/div/div[2]/p[2]/b'),
        'Dividend Yield': get_value_by_xpath('/html/body/div[4]/section[2]/div/div[3]/p[2]/b'),
        'Patrimônio Líquido': get_value_by_xpath('/html/body/div[4]/section[2]/div/div[4]/p[2]/b'),
        'Valor Patrimonial': get_value_by_xpath('/html/body/div[4]/section[2]/div/div[5]/p[2]/b'),
        'Rentab. no mês': get_value_by_xpath('/html/body/div[4]/section[2]/div/div[6]/p[2]/b'),
        'P/VP': get_value_by_xpath('/html/body/div[4]/section[2]/div/div[7]/p[2]/b')
        }
    
    print('\n', fii_data)

    data.append(fii_data)
    
    print(f'{((index + 1) * 100) / len(fii_links)}% - fii: {get_value_by_xpath('//*[@id="tickerName"]')}')
```

```python
fiis_data = pd.DataFrame(data)
```
